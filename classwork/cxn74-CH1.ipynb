```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 100, 1000)
y = np.linspace(1, 100, 1000)
z = 1/x

plt.subplot(2, 2, 1)
plt.plot(x, z, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'b')

plt.subplot(2, 2, 3)
plt.scatter(x, y**2)

plt.subplot(2, 2, 4)
plt.hist(x, 5)

plt.show()
```
