```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def piApproximation(N): # function to approximate pi with Monte Carlo method, from lectures slides
    count = 0
    for i in range(0, N):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        z = x*x + y*y
    
        if (z<1):
            count += 1
        
    pi_approx = 4 * count/N
    return pi_approx

# creating arrays to hold values of absolute and relative errors
absError = np.zeros(100)
relError = np.zeros(100)

for i in range(0, 100): # running function 100 times
    piApprox = piApproximation(1000) # N = 1000 samples
    absError[i] = abs(np.pi - piApprox)
    relError[i] = abs(np.pi - piApprox) / np.pi
         
plt.plot(absError, 'r-', label='absolute error')
plt.plot(relError, 'b-', label='relative error')
plt.legend()
plt.title('Absolute vs Relative Error of Monte Carlo Pi Approximation')
plt.show()
```
