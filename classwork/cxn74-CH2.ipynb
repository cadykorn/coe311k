```python
import numpy as np
```

```python
N = input("Enter number to convert to binary: ")
n = int(N)
q = 0
r = 0

print("Division\tQuotient\tRemainder")
b = []

while n > 0:
    q = int(n/2)
    r = n%2
    print('{:>4}{:<8}{:^16}{:^15}'.format(n, ' / 2', q, r))
    
    n = q
    b.append(r)

binary = np.array(b)
print(N,"in base-10 =", np.flip(binary), "in binary.")
```
