```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Part 1

# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

print("X:\n", X)
print("Y:\n", Y)
print("A:\n", A)

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"\nSlope: {m}, Intercept: {c}")
```

```python
y = 0.95*X + 0.59

plt.scatter(X,Y)
plt.plot(X, y)
plt.show()
```

```python
# Part 2

a = np.random.uniform(0, 10, 10)
b = np.random.uniform(0, 10, 10)
a.sort()

d = np.vstack([a, np.ones(len(a))]).T

m, c = np.linalg.lstsq(d, b, rcond=None)[0]

y1 = m*a + c

plt.scatter(a,b)
plt.plot(a, y1)
plt.show()
```

```python
# Part 3

x = np.arange(0, 1, 0.01)
a = np.random.uniform(-1, 1, len(x))

A = np.vstack([x, np.ones(len(x))]).T

y = (3/2)*x + 0.7 + a


m, c = np.linalg.lstsq(A, y, rcond=None)[0]

y1 = m*x + c

plt.scatter(x, y)
plt.plot(x, y1)
plt.show()
```
