```python
import numpy as np
import time
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
    
    
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def diagonallyDominant(A):
    print("A:\n", A)
    print("\nIs the matrix diagonally dominant?")
    
    for i in range (0, len(A)): # i represents the row number; diagonal element where row num = column num
        
        if np.abs(A[i,i]) < np.sum(np.abs(A[i])) - np.abs(A[i,i]):  
                        # sum of elements in row i - diagonal

            return False # if the abs value of the diagonal element is less than the absolute value of the sum of other elements
                         # in the row, the matrix is not diagonally dominant
        
    return True # otherwise, the diagonal element was never less than the sum of the other elements,
                # so the matrix IS diagonally dominant
```
