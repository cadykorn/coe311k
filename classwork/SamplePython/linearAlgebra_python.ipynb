```python
import numpy as np
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
```

```python
print(v1)
print(v2)
```

```python
v_add = v1 + v2
v_sub = v1 - v2
```

```python
print("v1 + v2 =", v_add)
print("v1 - v2 =", v_sub)
```

```python
scalar = 2
v_scale = (scalar * v1)
```

```python
print("v1 * 2 =", v_scale)
```

```python
dot_product = np.dot(v1, v2)
print("v1 * v2 =", dot_product)
```

```python
magnitude_v1 = np.linalg.norm(v1)
magnitude_v2 = np.linalg.norm(v2)

print(magnitude_v1)
print(magnitude_v2)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8, 9], [10, 11, 12], [13, 14, 15]])

print(A, "\n")
print(B)
```

```python
C = np.array([[1, 2, 3], [4, 5, 6]])
matrix_add = A + C
matrix_sub = A - C

print("A + C =\n", matrix_add)
print("A - C =\n", matrix_sub)
```

```python
transpose_A = A.T

print(A, "\n")
print(transpose_A)
```

```python
A = np.array([[2, 1], [1, 2]])
e_value, e_vector = np.linalg.eig(A)
print("eiganvalue:", e_value)
print("eigenvector:\n", e_vector)
```

```python
A = np.array([[4, 7], [2,6]])
A_inverse = np.linalg.inv(A)
print(A, "\n")
print(A_inverse)
```

```python
identity = np.dot(A, A_inverse)
print(identity)
```
