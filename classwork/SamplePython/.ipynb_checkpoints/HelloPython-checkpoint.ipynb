```python
import math

def calc_hyp(a, b): # function
    h = math.sqrt(a*a + b*b)
    return h
```

```python
my_h = calc_hyp(3, 4) # calling function
print(my_h)
```

```python
import random
import math

def approximatePi(numSamples):
    counter = 0
    for int in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        z = x*x + y*y
        if(z<1):
            counter+=1
    piApprox = 4 * counter / numSamples
    return piApprox
```

```python
myPiApprox = approximatePi(50)
print(myPiApprox)
```

```python
myPiApprox2 = approximatePi(1000)
print(myPiApprox2)
```
