```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50) # start, end (inclusive), amount of slices
print(x)
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), "r--", x, 2*np.sin(x), "b--")
plt.show()
```
