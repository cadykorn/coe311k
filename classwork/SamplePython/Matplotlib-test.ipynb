```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50) # start, end (inclusive), amount of slices
print(x)
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), "r--", x, 2*np.sin(x), "b--")
plt.show()
```

```python
plt.subplot(2, 1, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'g')
plt.show()
```

```python
plt.subplot(2, 2, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), 'b')
plt.subplot(2, 2, 4)
plt.plot(x, np.arcsin(x))
plt.show()
```

```python
def f1(x,y): # exact same output as previous code
    return (x*x+y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(45, 0)
```
