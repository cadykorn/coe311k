Taylor Series, Truncation Error

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def taylorSeriesApprox(N, x): # N is the number of terms, x is the value of sin being approximated
    sinApprox = 0
    for i in range(0, N):
        sinApprox += ((-1)**i) * (x**(2*i + 1)) / math.factorial(2*i + 1)
    return sinApprox

x = np.arange(0, 10, 0.1)

oneTerm = np.abs(np.sin(x) - np.array(taylorSeriesApprox(1, x)))
twoTerm = np.abs(np.sin(x) - np.array(taylorSeriesApprox(2, x)))
threeTerm = np.abs(np.sin(x) - np.array(taylorSeriesApprox(3, x)))
fourTerm = np.abs(np.sin(x) - np.array(taylorSeriesApprox(4, x)))
```

```python
plt.plot(x, oneTerm, 'r-', label='one term')
plt.plot(x, twoTerm, 'y-', label='two terms')
plt.plot(x, threeTerm,'g-', label='three terms')
plt.plot(x, fourTerm, 'b-', label='four terms')
plt.legend()
plt.xlabel('input value (x-value)')
plt.ylabel('absolute error')
plt.show()
```

output analysis: As I include more terms in the approximation, the error decreases at different rates for different x-values; smaller x-values start off with a smaller error, and the error difference between the varying number of terms increases at an increasing rate as the x-value increases (exponential, not uniform increase). Adding more terms would eventually decrease the error for larger x-values, while the smaller x-values are sufficiently approximated with less terms since the sin value is already relatively small.
