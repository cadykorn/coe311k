Monte Carlo Pi Approximation

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def piApproximation(N): # function to approximate pi with Monte Carlo method, from lecture slides
    count = 0
    for i in range(0, N):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        z = x*x + y*y
    
        if (z<1):
            count += 1
        
    pi_approx = 4 * count/N
    return pi_approx
    
x = np.arange(10, 10001, 20) # creating an array for the number of samples, previous knowledge from last semester COE301 course
y = np.zeros(np.size(x)) # creating an array for values of the error to replace
z = np.zeros(np.size(x))

for i in range(0, np.size(x)): # running the function for each number of samples
    y[i] = abs(np.pi - piApproximation(x[i])) # array now holds error values


plt.plot(x, y) # graph/plotting information from lecture slides
plt.title('Monte Carlo Pi Approximation Accuracy')
plt.xlabel('Sample Size')
plt.ylabel('Error')
plt.show()
```
