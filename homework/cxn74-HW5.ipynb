Least Squares Method

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def leastSquaresMethod(x, y, n):
    xsum = 0
    ysum = 0
    xxsum = 0
    xysum = 0

    for i in range(n):
        xsum += x[i]
        ysum += y[i]
        xxsum += (x[i]**2)
        xysum += (x[i] * y[i])

    m = ((n * xysum) - (xsum * ysum)) / (n * xxsum - (xsum**2))
    c = (ysum - m * xsum) / n
    
    return m, c

x = np.random.uniform(0, 10, 10)
x.sort()
y = np.random.uniform(0, 10, 10)
n = len(x)

m, c = leastSquaresMethod(x, y, n)
y1 = m*x + c
```

```python
plt.scatter(x,y)
plt.plot(x, y1)
plt.title("Custom Least Squares Method Plot with Random Vectors X and Y")
plt.show()
```

```python
x = np.arange(0, 1, 0.01)
n = len(x)
a = np.random.uniform(-1, 1, n)
y = (3/5*x) + 2.5 + a

m, c = leastSquaresMethod(x, y, n)
print("m =", m, "\nc =", c)
y1 = m*x + c
```

```python
print("Residuals:\n")

# largest residual
diffLarge = 0
for i in range (len(x)):
    diff = np.abs(y[i] - y1[i])
    print(diff)
    if (diff > diffLarge):
        diffLarge = diff
        xLarge = x[i]
        yLarge = y[i]
    
# smallest residual error
diffSmall = 100
for i in range (len(x)):
    diff = np.abs(y[i] - y1[i])
    if (diff < diffSmall):
        diffSmall = diff
        xSmall = x[i]
        ySmall = y[i]
        
print("\nPoint with largest residual: (", xLarge, ",", yLarge, ")")
print("Residual error =", diffLarge)

print("\n")

print("Point with smallest residual: (", xSmall, ",", ySmall, ")")
print("Residual error =", diffSmall)
```

```python
plt.scatter(x, y)
plt.plot(x, y1)
plt.title('Custom Least Square Method Plot')
plt.show()
```

```python
A = np.vstack([x, np.ones(len(x))]).T

m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print("m =", m, "\nc =", c)

y2 = m*x + c

plt.scatter(x, y)
plt.plot(x, y2)
plt.title("Numpy Least Square Method Plot")
plt.show()
```
